#Python

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress
```

```python
# Données
data = {
    'Epreuve_A': [14, 13, 16, 15, 18, 10, 12, 11, 15, 17, 16, 19],
    'Epreuve_B': [16, 14, 15, 18, 20, 13, 12, 14, 17, 18, 16, 20]
}
df = pd.DataFrame(data)
```

# Partie A
# 1. Tracer le nuage de points

```python

plt.scatter(df['Epreuve_A'], df['Epreuve_B'])
plt.xlabel('Épreuve A')
plt.ylabel('Épreuve B')
plt.title('Nuage de points Épreuve A vs Épreuve B')
plt.show()
```

# 2. Caractéristiques statistiques de chaque variable

```python

stats = df.describe().loc[['mean', '50%', 'min', 'max']]
stats.index = ['Moyenne', 'Médiane', 'Min', 'Max']
print("2. Caractéristiques statistiques :\n", stats)
```

# 3. Test de dépendance avec alpha = 1%

```python
slope, intercept, r_value, p_value, std_err = linregress(df['Epreuve_A'], df['Epreuve_B'])
dependance = p_value < 0.01
print("3. Dépendance entre les variables (alpha=1%) :", "Oui" if dependance else "Non")
```

# 4. Conclusion sur la dépendance

```python
conclusion = "Les variables sont dépendantes." if dependance else "Les variables ne sont pas dépendantes."
print("4. Conclusion :", conclusion)
```
